```python
# for webscrabing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
url="https://www.hubertiming.com/results/2018MLK"       #open url
html=urlopen(url)
```

```python
soup=BeautifulSoup(html)
```

```python
title=soup.title
print(title)
print(title.text)
```

```python
links=soup.find_all('a',href=True)
for link in links:
    print(link['href'])
```

```python
data=[]
allows=soup.find_all('tr')
print(allows[:10])
```

```python
for row in allows:
    row_list=row.find_all('td')
print(row_list)
```

```python
for row in allows:
    row_list=row.find_all('td')
    dataRow=[]
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)
print(data)
```

```python
for row in allows:
    row_list=row.find_all('td')
    dataRow=[]
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)

titles=data[0]
data=data[1:]
print(titles)
print(data[-2:])
```

```python
for row in allows:
    row_list=row.find_all('td')
    dataRow=[]
    for cell in row_list:
        dataRow.append(cell.text)
    data.append(dataRow)
data=data[1:]
print(data[-2:])
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
print(df.head())
print(df.tail(2))
```
