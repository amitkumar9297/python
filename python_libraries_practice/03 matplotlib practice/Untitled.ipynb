# Other 2D Graph

```python
n=np.array([0,1,2,3,4,5])

fig, axes=plt.subplots(1,4,figsize=(16,5))

axes[0].set_title("scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)

axes[2].set_title("bar")
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)

axes[3].set_title("fill between")
axes[3].fill_between(x,x**2,x,x**3,color="green",alpha=0.5)
```

```python
"""draw a radar chart"""

fig =plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,1,1],polar=True)

t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*2,color="blue",lw=3)
```

```python
"""draw a histogram"""

n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("default histogram")
axes[0].hist(n)

axes[1].set_title("cumulative detailed histogram")
axes[1].hist(n,cumulative=True,bins=10)
```

```python
# draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2 - Y**2)
Z2=np.exp(-(X-1)**2 - (Y-1)**2)
Z=(Z1-Z2)*2

print(X)
print(Y)
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title("contour map")
```

# write a python programm to create a pie chart of popularity of programming language


```python
import matplotlib.pyplot as plt

#Data to plot
languages='python','java','php','javascript','c#','c++'
popularity=[22.2,17.6,8.8,8,7.7,6.7]
colors=["#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd","#8c564b"]
```

```python
# explod 1st slice

explode=(0.1, 0, 0, 0, 0, 0)
plt.pie(popularity,explode=explode,labels=languages,colors=colors,autopct='%1.1f%%',shadow=True,startangle=140)
```

```python
explode=(0.1, 0, 0, 0, 0, 0)
plt.pie(popularity,explode=explode,labels=languages,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
```
